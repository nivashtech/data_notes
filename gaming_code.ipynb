```python
import random
def word_guessing_game():
    word_list=["table","chair","fan","printer","laptop","sofa","remote"]
    secreat_word=random.choice( word_list)
    #print(secreat_word)
    length=len(secreat_word)
    print("welcome to word guessing game")
    print(f"the length of secreat word is :{length}")
    print(f"you have only {length} attempt")
    for i in range(0,length):
        user_input=input("enter your guessing letter which is in secreat word")
        if user_input in secreat_word:
            count=secreat_word.count(user_input)
            posithion=secreat_word.find(user_input)
            print(f"weldon your {user_input} is in that word and have {count} times and {posithion} th position")
            #continue
        
        elif user_input not in secreat_word:
            print("your guessing letter not in secreat word")
    print(f"you done all atempts well finally the word is {secreat_word}")
   
while True:
    word_guessing_game()
    replay=input(" are you want continue this game")
    if replay=="yes":
        continue
    elif replay=="no":
        print("thank you, game is end")
        break
    
        
    
    
```
